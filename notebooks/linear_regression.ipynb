```python
import sys
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt

from models import LinearRegression
```

```python
data = pd.read_csv('../data/stackloss.csv')

for col in data:
    if col == 'StackLoss':  
        continue
    data['{}_sq'.format(col)] = data[col]**2

X = data[[col for col in data.columns if col != 'StackLoss']].values
y = data.StackLoss.values

lr = LinearRegression(X = X, y = y)
lr.fit()

ypred = lr.predict(X)

fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_size_inches(10,5)

ax1.plot(y, label = 'True')
ax1.plot(ypred, label = 'Predicted')

ax1.set_title('True vs Predicted')
ax1.legend()

ax2.scatter(ypred, y)
```
